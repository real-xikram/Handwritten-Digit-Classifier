# 1. Import the libraries

```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from tensorflow.keras.layers import Dense , Dropout , Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from keras import backend as K
import tensorflow as tf
```

# Load the data

```python
(x_train , y_train) , (x_test , y_test) = mnist.load_data()
print(x_train.shape , y_train.shape)
```

```python
y_train
```

# Preprocess the data

```python
x_train = x_train.reshape( x_train.shape[0] , 28 ,28, 1)
x_test = x_test.reshape(x_test.shape[0] , 28 , 28 , 1)
input_shape = ( 28 , 28 , 1)

#tf.keras.utils.to_categorical does one hot encoding for each class
y_train = keras.utils.to_categorical(y_train , None)
y_test = keras.utils.to_categorical(y_test , None)

x_train = x_train.astype('float32')
x_test = x_test.astype('float')
```

```python
x_train
```

```python
#ensuring all vaues are between 0-255
x_train /= 255
x_test /= 225

print('x_train shape: ', x_train.shape)
print(x_train.shape[0],  'train samples')
print(x_test.shape[0], 'test samples')
```

# Create a model

```python
batch_size = 128
num_classes = 10
epochs = 20
model = tf.keras.Sequential()
model.add( Conv2D( 32 , kernel_size= (3 ,3)  , activation= 'relu' , input_shape = input_shape))
model.add( Conv2D( 64 , (3,3) , activation= 'relu' ))
model.add( MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.25) )
model.add( Flatten() )
model.add( Dense ( 256 , activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes , activation= 'softmax'))

model.compile( loss = keras.losses.categorical_crossentropy , optimizer = tf.keras.optimizers.Adadelta() , metrics = ['accuracy'])
```

# Train the model

```python
hist = model.fit(x_train , y_train , batch_size= batch_size , epochs= epochs , verbose= 1 , validation_data= (x_test , y_test))
print("The model has successfully trained")

model.save('mnist.h5')
print("Saving the model as mnsit.h5")
```

# Evaluate the model

```python
score = model.evaluate(x_test , y_test , verbose= 0)
print('Test loss :' , score[0])
print('Test Accuracy :' , score[1])
```
